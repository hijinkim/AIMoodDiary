```python
sentence.info()
```

## morpheme analysis with khaiii

```python
from khaiii import KhaiiiApi
api = KhaiiiApi()
```

```python
def morphs_khaiii(text):
  for word in api.analyze(text):
    for morph in word.morphs:
      if morph.tag in ['NNG', 'NNP']:
        return morph.lex
```

```python
result = []
for s in sentence['original']:
  print(s)
  result.append(morphs_khaiii(s))
```

```python
result
```

```python
result = [r for r in result if r]
```

```python
result
```

## frequency analysis with nltk

```python
ko = nltk.Text(result, name = 'test')

plt.figure(figsize=(16,8))
ko.plot(50)
plt.show()
```

```python
ko.vocab()
```

```python
vocab_counts = ko.vocab()
```

```python
vocab_dict = dict(vocab_counts)
```

```python
vocab_dict_sorted = sorted(vocab_dict.items(), reverse=True, key=lambda item:item[1])
```

## wordcloud

```python
from wordcloud import WordCloud

data = ko.vocab().most_common(100)

path = '/usr/share/fonts/truetype/nanum/NanumGothicExtraBold.ttf'
wc = WordCloud(font_path=path, background_color='white', width=1500, height=1000).generate_from_frequencies(dict(data))

wc.to_file('wordcloud_khaiii.jpg')
```

```python
plt.figure(figsize=(16, 9))
plt.imshow(wc)
```
