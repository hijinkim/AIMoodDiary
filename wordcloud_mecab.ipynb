## load data

```python
import pandas as pd

DF = pd.read_excel('preprocessing_sad.xlsx', engine='openpyxl')

DF.info()
```

```python
sentence = pd.DataFrame()
```

```python
sentence['original'] = DF['final']
```

```python
sentence.info()
```

## morpheme analysis with mecab

```python
from konlpy.tag import Mecab

t= Mecab()

def morphs(text):
  tokens = t.pos(text)
  for i in tokens:
    if i[1] in ['NNG', 'NNP']: ## 고유명사와 일반명사만 return
      return i[0]
```

```python
result = []
for s in sentence['original']:
  result.append(morphs(s))
```

```python
result
```

```python
result = [r for r in result if r]
```

```python
result
```

## frequency analysis with nltk

```python
import nltk
from konlpy.corpus import kobill
```

```python
ko = nltk.Text(result, name = 'wordcloud')

plt.figure(figsize=(16,8))
ko.plot(50)
plt.show()
```

## wordcloud

```python
from wordcloud import WordCloud

data = ko.vocab().most_common(100)

path = '/usr/share/fonts/truetype/nanum/NanumGothicExtraBold.ttf'
wc = WordCloud(font_path=path, relative_scaling=0.8,
               background_color='white').generate_from_frequencies(dict(data))

plt.figure(figsize=(16, 8))
plt.imshow(wc)
plt.axis('off')
plt.show
```
