# mecab

## load data

```python
import pandas as pd

DF = pd.read_excel('preprocessing_sad.xlsx', engine='openpyxl')

DF.info()
```

```python
sentence = pd.DataFrame()
```

```python
sentence['original'] = DF['final']
```

```python
sentence.info()
```

## morpheme analysis with mecab

```python
from konlpy.tag import Mecab

t= Mecab()

def morphs(text):
  tokens = t.pos(text)
  for i in tokens:
    if i[1] in ['NNG', 'NNP']: ## 고유명사와 일반명사만 return
      return i[0]
```

```python
result_mecab = []
for s in sentence['original']:
  result_mecab.append(morphs(s))
```

```python
result_mecab = [r for r in result_mecab if r]
```

```python
result_mecab
```

## frequency analysis with nltk

```python
import nltk
from konlpy.corpus import kobill
```

```python
ko_mecab = nltk.Text(result_mecab, name = 'wordcloud')

plt.figure(figsize=(16,8))
ko_mecab.plot(50)
plt.show()
```

## wordcloud

```python
from wordcloud import WordCloud

data = ko_mecab.vocab().most_common(100)

path = '/usr/share/fonts/truetype/nanum/NanumGothicExtraBold.ttf'
wc = WordCloud(font_path=path, relative_scaling=0.8,
               background_color='white').generate_from_frequencies(dict(data))

plt.figure(figsize=(16, 8))
plt.imshow(wc)
plt.axis('off')
plt.show
```

# khaiii

## morpheme analysis with khaiii

```python
from khaiii import KhaiiiApi
api = KhaiiiApi()
```

```python
def morphs_khaiii(text):
  for word in api.analyze(text):
    for morph in word.morphs:
      if morph.tag in ['NNG', 'NNP']:
        return morph.lex
```

```python
result_kha = []
for s in sentence['original']:
  result_kha.append(morphs_khaiii(s))
```

```python
result_kha = [r for r in result_kha if r]
```
